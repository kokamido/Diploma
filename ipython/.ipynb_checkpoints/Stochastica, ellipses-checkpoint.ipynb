```python
p = 1
Step = 0.001
q = 0.1
```

```python
def f(x,y):
    return 1-x*y
```

```python
def g(x,y):
    return p*y*(x-(1+q)/(q+y))
```

```python
def k1(x0,y0):
    return Step*f(x0,y0)
```

```python
def l1(x0,y0):
    return Step*g(x0,y0)
```

```python
def k2(x0,y0):
    x = x0+k1(x0,y0)/2
    y = y0+l1(x0,y0)/2 
    return Step*f(x,y)
```

```python
def l2(x0,y0):
    x = x0+k1(x0,y0)/2
    y = y0+l1(x0,y0)/2
    return Step*g(x,y)
```

```python
def k3(x0,y0):
    x = x0+k2(x0,y0)/2
    y = y0+l2(x0,y0)/2
    return Step*f(x,y)
```

```python
def l3(x0,y0):
    x = x0+k2(x0,y0)/2
    y = y0+l2(x0,y0)/2
    return Step*g(x,y)
```

```python
def k4(x0,y0):
    x = x0+k3(x0,y0)
    y = y0+l3(x0,y0)
    return Step*f(x,y)
```

```python
def l4(x0,y0):
    x = x0+k3(x0,y0)
    y = y0+l3(x0,y0)
    return Step*g(x,y)
```

```python
def rk(pair):
    x0 = float(pair[0])
    y0 = float(pair[1])
    x = pair[0] +1/6*(k1(x0,y0)+2*k2(x0,y0)+2*k3(x0,y0)+k4(x0,y0))
    y = pair[1] +1/6*(l1(x0,y0)+2*l2(x0,y0)+2*l3(x0,y0)+l4(x0,y0))
    return [x,y]
```

```python
def curve(x0,y0,n):
    i = 0
    res_x = list()
    res_y = list()
    while i < n:
        res_x.append(x0)
        res_y.append(y0)
        current = rk([x0,y0])
        x0 = current[0]
        y0 = current[1]
        i+=1
    return[res_x,res_y]
```

```python
def curve_reverse(x0,y0,n):
    global Step
    buf = Step
    Step = -buf
    i = 0
    res_x = list()
    res_y = list()
    while i < n:
        res_x.append(x0)
        res_y.append(y0)
        current = rk([x0,y0])
        x0 = current[0]
        y0 = current[1]
        i+=1
    Step = buf
    return[res_x,res_y]
```

```python
def curves_y(start_y,end_y,x,step,rk_iter,color='c',lw=1):
    y = start_y
    while y < end_y:
        current_curve = curve(x,y,rk_iter)
        plt.plot(current_curve[0],current_curve[1],color,linewidth=lw)
        y+=step
```

```python
def curves_x(start_x,end_x,y,step,rk_iter):
    x = start_x
    while x < end_x:
        current_curve = curve(x,y,rk_iter)
        plt.plot(current_curve[0],current_curve[1],'c',linewidth=5)
        x+=step
```

```python
def curves_circle(num, r,rk_iter, color='c',lw=1):
    for i in range(1,num):
        current_curve = curve(1+r*math.cos(2*math.pi/num*i),1+r*math.sin(2*math.pi/num*i),rk_iter)
        plt.plot(current_curve[0],current_curve[1],color,linewidth=lw)
```

```python
def curves_circle_rev(num, r,rk_iter, color='c',lw=1):
    for i in range(1,num):
        current_curve = curve_reverse(1+r*math.cos(2*math.pi/num*i),1+r*math.sin(2*math.pi/num*i),rk_iter)
        plt.plot(current_curve[0],current_curve[1],color,linewidth=lw)
```

```python
import random
import math
random.seed()
def normal_dist_random_values():
    x = random.random()
    y = random.random()
    
    a = math.sqrt(-2*math.log(x))*math.cos(2*math.pi*y)
    b = math.sqrt(-2*math.log(x))*math.sin(2*math.pi*y)
    return [a,b]
noise_lvl = 0.01

def noisy_curve(x0,y0,n):
    i = 0
    res_x = list()
    res_y = list()
    print(noise_lvl)
    while i < n:
        #if i%100000 == 0:
        #    print(i/100000)
        res_x.append(x0)
        res_y.append(y0)
        current = rk([x0,y0])
        noise = normal_dist_random_values()
        x0 = current[0]+noise_lvl*noise[0]*math.sqrt(Step)
        y0 = current[1]+noise_lvl*noise[1]*math.sqrt(Step)
        i+=1
    return[res_x,res_y]

def avg(list_arg):
    return sum(list_arg)/len(list_arg)

def expectancy(crv):
    return [avg(crv[0]),avg(crv[1])]

from functools import reduce
def dispersion(crv):
    exp = expectancy(crv)
    centr_curve = [map(lambda x: x-exp[0],crv[0]), map(lambda x: x-exp[1],crv[1])]
    return [reduce(lambda y,x: x*x+y,centr_curve[0])/len(crv[0]),reduce(lambda y,x: x*x+y,centr_curve[1])/len(crv[0])]
```

```python
def normalize(vec):
    mod = euclidean_range(0,0,vec[0],vec[1])
    return[vec[0]/mod,vec[1]/mod]
def time_row(vals):
    times = [i*Step for i in range(0,len(vals))]
    plt.plot(times,vals)
def get_W_matrix():
    w11 = (-p*p-p*q*q-p*q-q*q-2*q-1)/(2*q*p*(p-q-1))
    w12=w21=(p*p*q+p*p+q*q+2*q+1)/(2*q*p*(p-q-1))
    w22=-((q+1)*(p*p*q+p*p+p*q+q+1))/(2*q*p*(p-q-1))
    return[[w11,w12],[w21,w22]]
```

```python
def find_z_multiplier(p):
    k = math.sqrt(-math.log(1-p))
    W = get_W_matrix()
    e_values = eugenvalues(W)
    e_vectors = eigenvectors(W)
    return[math.sqrt(2*e_values[0])*noise_lvl*k,math.sqrt(2*e_values[1])*noise_lvl*k]

def z(phi):
    mul = find_z_multiplier(0.99)
    return [mul[0]*math.cos(phi), mul[1]*math.sin(phi)]

def get_delta():
    vec = eigenvectors(get_W_matrix())
    return vec[0][0]*vec[1][1]-vec[0][1]*vec[1][0]

def x(phi):
    delta = get_delta()
    e_vec = eigenvectors(get_W_matrix())
    z_2 = z(phi)[0]
    z_1 = z(phi)[1]
    x_1 = 1+(z_1*e_vec[1][1]-z_2*e_vec[0][1])/delta
    x_2 = 1+(z_2*e_vec[0][0]-z_1*e_vec[1][0])/delta
    return [x_1,x_2]

def get_ellipse():
    x_1 = list()
    x_2 = list()
    for i in range (0,361):
        buf_x = x(2*math.pi/360*i)
        x_1.append(buf_x[0])
        x_2.append(buf_x[1])
    return[x_1,x_2]
```
